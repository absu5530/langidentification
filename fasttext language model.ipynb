```python
sents.columns = ['index', 'lang', 'text']
```

```python
samples_ind = pd.DataFrame()
```

```python
langs_list = []
texts_list = []
for subdir, dirs, files in os.walk('/Users/uabhifn/Downloads/dakshina_dataset_v1.0/'):
    for file in files:
        if 'roman.' in file:
            with open(os.path.join(subdir, file)) as f:
                texts = f.readlines()
            texts_list.extend([t.strip() for t in texts])
            langs_list.extend([file.split('.')[0] + '-rom'] * len(texts))
```

```python
samples_ind['lang'] = langs_list
samples_ind['text'] = texts_list
```

```python
samples_ind
```

```python
samples_ind['lang'].value_counts()
```

```python
egy = pd.read_csv('/Users/uabhifn/Downloads/Arabizi Identification/arabizi-twitter-egy.csv')
leb = pd.read_csv('/Users/uabhifn/Downloads/Arabizi Identification/arabizi-twitter-leb.csv')
```

```python
tun = pd.read_csv('~/Downloads/tunizi_train')
```

```python
samples_ar = pd.DataFrame()
```

```python
samples_ar['text'] = pd.concat([tun['text'], egy[egy['arabizi'] == '1']['tweet_filter'], leb[leb['arabizi'] == '1']['tweet_filter']])
```

```python
samples_ar['lang'] = 'ar-rom'
```

```python
samples_ar
```

```python
samples_rom = pd.concat([samples_ind, samples_ar])
```

```python
augmented_data = pd.concat([sents[['lang', 'text']], samples_rom])
```

```python
augmented_data
```

```python
augmented_data.reset_index(inplace=True)
augmented_data['index'] = augmented_data.index
```

```python
augmented_data.to_csv('sentences_augmented.csv', header=False, sep='\t', index=False)
```
